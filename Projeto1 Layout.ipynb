```python
dados=pd.read_csv('DOM2015csv.zip',sep=',')
```

```python
dados.dtypes

#VO2322 = acesso à internet no próprio domicílio
```

```python
#região ():
#dados=dados[dados.UF==35]

#categorizando dados de acesso à internet:

dados.V02322=dados.V02322.astype("category") #acesso no próprio domicílio
dados.V02322.cat.categories=(['Sim', 'Não'])

dados.V02323=dados.V02323.astype("category") #acesso por microcomputador
dados.V02323.cat.categories=(['Sim', 'Não'])

dados.V02324=dados.V02324.astype("category") #acesso por telefone
dados.V02324.cat.categories=(['Sim', 'Não'])

dados.V02325=dados.V02325.astype("category") #acesso por tablet
dados.V02325.cat.categories=(['Sim', 'Não'])

dados.V02326=dados.V02326.astype("category") #acesso pela TV
dados.V02326.cat.categories=(['Sim', 'Não'])

dados.V02327=dados.V02327.astype("category") #acesso por outras formas
dados.V02327.cat.categories=(['Sim', 'Não'])

dados.V02424=dados.V02424.astype("category") #acesso através de rede celular 
dados.V02424.cat.categories=(['Sim', 'Não'])

dados.V02425=dados.V02425.astype("category") #acesso por linha telefônica
dados.V02425.cat.categories=(['Sim', 'Não'])

dados.V02426=dados.V02426.astype("category") #acesso de banda larga 
dados.V02426.cat.categories=(['Sim', 'Não'])

#área:
dados.V4105=dados.V4105.astype("category")
dados.V4105.cat.categories=(['URBANO_1', 'URBANO_2', 'URBANO_3', 'RURAL_1', 'RURAL_2', 'RURAL_3', 'RURAL_4', 'RURAL_5'])

dados.loc[(dados.V4105=='URBANO_1') | (dados.V4105=='URBANO_2') | (dados.V4105=='URBANO_3'), 'Área']='Urbano'
dados.loc[(dados.V4105=='RURAL_1') | (dados.V4105=='RURAL_2') | (dados.V4105=='RURAL_3') | (dados.V4105=='RURAL_4') | (dados.V4105=='RURAL_5'), 'Área']='Rural'

#Criando a lista filtrada

dataframe_filtro=pd.DataFrame()

tempData = dados[(dados.V02426.notnull()) & (dados.V4105.notnull()) & (dados.V02425.notnull()) & (dados.V02424.notnull()) & (dados.V02327.notnull()) & (dados.V02326.notnull()) & (dados.V02325.notnull()) & (dados.V02324.notnull())  & (dados.V02323.notnull())  & (dados.V02322.notnull())]

dataframe_filtro["acesso_de_banda_larga"]=tempData.V02426

dataframe_filtro.loc[(dados.V02426=='Sim') | (dados.V02425=='Sim') | (dados.V02424=='Sim') | (dados.V02327=='Sim') | (dados.V02326=='Sim') | (dados.V02325=='Sim') | (dados.V02324=='Sim') | (dados.V02323=='Sim') | (dados.V02322=='Sim'), 'Acesso à internet']='possuí'
dataframe_filtro.loc[(dados.V02426=='Não') & (dados.V02425=='Não') &  (dados.V02424=='Não') &  (dados.V02327=='Não') &  (dados.V02326=='Não') &  (dados.V02325=='Não') &  (dados.V02324=='Não') &  (dados.V02323=='Não') &  (dados.V02322=='Não'), 'Acesso à internet']='não possuí'

dataframe_filtro.loc[(dados.V4105=='URBANO_1') | (dados.V4105=='URBANO_2') | (dados.V4105=='URBANO_3'), 'Área']='Urbano'
dataframe_filtro.loc[(dados.V4105=='RURAL_1') | (dados.V4105=='RURAL_2') | (dados.V4105=='RURAL_3') | (dados.V4105=='RURAL_4') | (dados.V4105=='RURAL_5'), 'Área']='Rural'


dataframe_filtro

dataframe_filtro.dtypes

banda_larga=dataframe_filtro.acesso_de_banda_larga.value_counts(sort=False)

dataframe_filtro


```

```python
plot=banda_larga.plot(kind='pie', title='Acesso a banda larga',legend=("possuí","não possuí"),autopct='%.1f',figsize=(5,5))

#função plt.subplot serve para colocar vários gráficos
#na mesma imagem para fins de comporação.
```
